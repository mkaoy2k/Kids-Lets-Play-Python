```python
""" Examples for Sorting Lists
Two ways of sorting lists:
1. By sorted() function to create another sorted list
2. By sort() method to sort on the same list"""

#
li = [9, 1, 8, 2, 7, 3, 6, 4, 5]
print(f'Example of sorting list via sorted():\n')
print(f'===>Original list:\n{li}\n')
print(f'===>Type: {type(li)}\n')

# sorted() function returns a sorted list
# from the given original list (any iterable, e.g. immutable tuple)
s_li = sorted(li)
print(
    f'===>New sorted list created by sorted() function:\n{s_li}\n===>Type: {type(s_li)}\n')

# Another way to sort a list by invoking its associated method sort
li = [9, 1, 8, 2, 7, 3, 6, 4, 5]
print(f'Example of sorting list via sort():\n{li}\n')
print(f'===>Original list:\n)')
print(f'===>Type: {type(li)}')

li.sort()
print(f'===>Original List after sort() method:\n{li}\n')

# Sort in a reverse order
s_li = sorted(li, reverse=True)
print(f'===>Reversed sorted List by sorted() function:\n{s_li}\n')

li = [9, 1, 8, 2, 7, 3, 6, 4, 5]

li.sort(reverse=True)
print(f'Original List after involking reverse sort() method:\n{li}\n')

# Sort with absolute integer value
print(f'Example of sorting with key = absolute integer value')
li = [-6, -5, -4, 1, 2, 3]
print('Original List:\t', li)

s_li = sorted(li, key=abs)
print('   abs sorted List:\t', s_li)
```

```python
"""Examples of Sorting Tuples by sorted() function
Note: no sort method with Tuples,
since tuples are immutable
i.e. <tuple obj>.sort() resulted in an error"""

tup = (9, 1, 8, 2, 7, 3, 6, 4, 5)
print(f'Original Tuple:')
print(f'{tup}')
print(f'===>Type: {type(tup)}\n')

# Sort a tuple
print(f'Example of sorting a tuple via sorted() function:')
sl_tup = sorted(tup)
print(f'===>Resulted in a sorted list:\n{sl_tup}\n')
```

```python
""" Examples of Sorting Dictionaries
1. Sorted by key
2. Sorted by value, assuming sortable alphanumeric attribute values"""

# Original dictionary
di = {'name': 'Mike', 'job': 'programmer', 'age': '20', 'os': 'Mac'}
print(f'Original Dictionary:')
print(f'{di}')
print(f'===>Type: {type(di)}\n')

# Sort dictionary by key
print(f'Example of sorting dictionary by key via sorted() function:')
s_di = sorted(di)
print(f'===>Sorting dictionary by key, resulted in a list of sorted keys:')
print(f'{s_di}\n')

# print dictionary sorted by key
print(f'Example of sorting dictionary by key via sort() method:')
s_li = sorted(di.items(), key=lambda kv: (kv[0], kv[1]))
print(f'===>Resulted in a list of sorted tuples of kv-pairs:\n{s_li}\n')

# print dictionary sorted by value
print(f'Example of sorting dictionary by value via sort() method:')
s_li = sorted(di.items(), key=lambda kv: (kv[1], kv[0]))
print(f'===>Resulted in a list of sorted tuples of kv-pairs:\n{s_li}\n')
```

```python
import datetime

# 
class Person:
    
    #  
    def __init__(self, last, first, dob):
        self.last_name = last
        self.first_name = first
        dob_yyyy, dob_mm, dob_dd = dob.split('-')
        self.birth_date = datetime.date(int(dob_yyyy), 
                                        int(dob_mm), 
                                        int(dob_dd))
    #  email 
    @property
    def email(self):
        return f'{self.last_name}{self.first_name}@school.edu.tw'

    #  fullname  
    @property
    def fullname(self):
        return f'{self.last_name} {self.first_name}'
    
    #  email 
    @fullname.setter
    def fullname(self, name):
        last, first = name.split(' ')
        self.last_name = last
        self.first_name = first

# 
class Student(Person):

    #  
    def __init__(self, last, first, dob, class_id, courses=None):
        super().__init__(last, first, dob)
        self.class_id = class_id
        if courses is None:
            self.courses = {}
        else:
            self.courses = courses
    
    #  
    def add_classId(self, class_id):
        self.class_id = class_id
        
    #  
    def add_course(self, course, score=0):
        self.courses.update({course : score})

    #  
    def remove_course(self, course):
        if course in self.courses:
            self.courses.pop(course)
    
    #  
    def print_courses(self):
        print(self.fullname)
        for course, score in self.courses.item():
            print(f'--> {course}: {score}')

# 
student_1 = Student('', '', '2010-3-15', '6', {'':0, '':0})
student_2 = Student('', '', '2011-12-20', '6', {'':0, '':0, '':0})

print(f'{student_1.fullname}')
print(f'--->  = {student_1.birth_date}')
print(f'--->  = {student_1.email}')
print(f'--->  = {student_1.class_id}')
print(f'--->  = {student_1.courses}\n')


print(f'{student_2.fullname}')
print(f'--->  = {student_2.birth_date}')
print(f'--->  = {student_2.email}')
print(f'--->  = {student_2.class_id}')
print(f'--->  = {student_2.courses}\n')
```

```python
import datetime

# 
class Person:

    # 
    head_count = 0

    #  
    def __init__(self, last, first, dob):
        self.last_name = last
        self.first_name = first
        dob_yyyy, dob_mm, dob_dd = dob.split('-')
        self.birth_date = datetime.date(int(dob_yyyy),
                                        int(dob_mm),
                                        int(dob_dd))
        # 
        Person.head_count += 1

    #  email 

    @property
    def email(self):
        return f'{self.last_name}{self.first_name}@school.edu.tw'

    #  fullname  
    @property
    def fullname(self):
        return f'{self.last_name} {self.first_name}'

    #  email 
    @fullname.setter
    def fullname(self, name):
        last, first = name.split(' ')
        self.last_name = last
        self.first_name = first

# 
class Staff(Person):

    #  
    def __init__(self, last, first, dob, title, jobs=None):
        super().__init__(last, first, dob)
        self.title = title

        if jobs is None:
            self.jobs = []
        else:
            self.jobs = jobs


# 
staff_1 = Staff('', '', '1958-6-7', '', ['', ''])

print(f'{staff_1.fullname}')
print(f'---> {staff_1.title}')
print(f'---> {staff_1.jobs}\n')

#  (inheritance)
print(f': {staff_1.head_count} \n')

staff_2 = Staff('', '', '1988-9-9', '', ['', ''])

print(f'{staff_2.fullname}')
print(f'---> {staff_2.title}')
print(f'---> {staff_2.jobs}\n')

print(f': {Person.head_count} \n')
```
