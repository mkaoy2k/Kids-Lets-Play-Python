####  for 


```python
message = ', . !'

print('for ...')

for char in message:
    print(f'{char}', end='-')
print()
print('for ')
```

```python
# break statement
message = ', . !'

print('for ...')

for char in message:

    if char == ".":
        print(f'\n===> "{char}" !')
        
        #  break
        break
    
    print(f'{char}', end='-')

print('for \n')
```

```python
#  (continue statement)
print('for ...')

for char in message:
    
    if char == ".":
        print(f'\n===> "{char}" !')
        
        #  continue
        continue
    
    print(f'{char}', end='-')
print()
print('for ')
```
